```python
import pyspedas 
from pytplot import tplot, tplot_names, get_data, store_data
from pyspedas import mms_load_fgm, mms_load_fpi
import matplotlib.pyplot as plt
import numpy as np
import math as math
import pytplot as py
import statistics as stat
from pyspedas.mms import fpi
import shelve

```

```python
downstreamFgmData=[]
downstreamDensityData=[]
downstreamVelocityData=[]

upstreamFgmData=[]
upstreamDensityData=[]
upstreamVelocityData=[]
ratio = []
G_u=[]
G_d=[]
#try not to refresh
```

```python
#(downstream to upstream)
def eventAvg(trangeD,trangeU):
    pyspedas.mms.fgm(trange=trangeD, time_clip=True )
    fpi(datatype = 'dis-moms', trange=trangeD, center_measurement=True,  time_clip=True)
    #fpi(datatype='des-moms', trange=trangeD, center_measurement=True, time_clip=True)
    
    #fgm
    fgm_d = get_data('mms1_fgm_b_dmpa_srvy_l2')
    fgm_dx = [row[0] for row in fgm_d[1]]
    fgm_dy = [row[1] for row in fgm_d[1]]
    fgm_dz = [row[2] for row in fgm_d[1]]
    fgm_dx_avg = stat.mean(fgm_dx)
    fgm_dy_avg = stat.mean(fgm_dy)
    fgm_dz_avg = stat.mean(fgm_dz)
    fgm_d_total_avg = math.sqrt(fgm_dx_avg**2+fgm_dy_avg**2+fgm_dz_avg**2)
    downstreamFgmData.append(fgm_d_total_avg)
    arr_downstreamFgmData = np.array(downstreamFgmData)
    #print(fgm_d_total_avg)
    G_d.append(fgm_d_total_avg*1e-9)
    arr_G_d = np.array(G_d)
    
    #Density(dis)
    dDis_d = get_data('mms1_dis_numberdensity_fast')
    dDis_d_avg = stat.mean(dDis_d[1])
    downstreamDensityData.append(dDis_d_avg)
    arr_downstreamDensityData = np.array(downstreamDensityData)


    #print(dDis_d_avg)

   
    #velocity(dis)
    vDis_d = get_data('mms1_dis_bulkv_dbcs_fast')
    vDis_dx = [row[0] for row in vDis_d[1]]
    vDis_dy = [row[1] for row in vDis_d[1]]
    vDis_dz = [row[2] for row in vDis_d[1]]
    vDis_dx_avg = stat.mean(vDis_dx)
    vDis_dy_avg = stat.mean(vDis_dy)
    vDis_dz_avg = stat.mean(vDis_dz)
    vDis_d_total_avg = math.sqrt(vDis_dx_avg**2+vDis_dy_avg**2+vDis_dz_avg**2)
    downstreamVelocityData.append(vDis_d_total_avg)
    arr_downstreamVelocityData = np.array(downstreamVelocityData)

    
    #print(vDis_d_total_avg)
    
  
    

    pyspedas.mms.fgm(trange=trangeU, time_clip=True )
    fpi(datatype = 'dis-moms', trange=trangeU, center_measurement=True,  time_clip=True)
    fpi(datatype='des-moms', trange=trangeU, center_measurement=True, time_clip=True)
    
    #fgm
    fgm_u = get_data('mms1_fgm_b_dmpa_srvy_l2')
    fgm_ux = [row[0] for row in fgm_u[1]]
    fgm_uy = [row[1] for row in fgm_u[1]]
    fgm_uz = [row[2] for row in fgm_u[1]]
    fgm_ux_avg = stat.mean(fgm_ux)
    fgm_uy_avg = stat.mean(fgm_uy)
    fgm_uz_avg = stat.mean(fgm_uz)
    fgm_u_total_avg = math.sqrt(fgm_ux_avg**2+fgm_uy_avg**2+fgm_uz_avg**2)
    upstreamFgmData.append(fgm_u_total_avg)
    arr_upstreamFgmData = np.array(upstreamFgmData)
    G_u.append(fgm_d_total_avg*1e-9*10000)
    arr_G_u = np.array(G_u)
    


    #print(fgm_u_total_avg)

    #Density(dis)
    dDis_u = get_data('mms1_dis_numberdensity_fast')
    dDis_u_avg = stat.mean(dDis_u[1])
    upstreamDensityData.append(dDis_u_avg)
    arr_upstreamDensityData = np.array(upstreamDensityData)


    #print(dDis_u_avg)

    
    #velocity(dis)
    vDis_u = get_data('mms1_dis_bulkv_dbcs_fast')
    vDis_ux = [row[0] for row in vDis_u[1]]
    vDis_uy = [row[1] for row in vDis_u[1]]
    vDis_uz = [row[2] for row in vDis_u[1]]
    vDis_ux_avg = stat.mean(vDis_ux)
    vDis_uy_avg = stat.mean(vDis_uy)
    vDis_uz_avg = stat.mean(vDis_uz)
    vDis_u_total_avg = math.sqrt(vDis_ux_avg**2+vDis_uy_avg**2+vDis_uz_avg**2)
    upstreamVelocityData.append(vDis_u_total_avg)
    arr_upstreamVelocityData = np.array(upstreamVelocityData*1000)


    #print(vDis_u_total_avg)
    
    R = dDis_d_avg/dDis_u_avg
    ratio.append(R)
    arr_ratio = ratio
    s = shelve.open('eventData.db')
    s['fgmd'] = arr_downstreamFgmData    
    s['densityd'] = arr_downstreamDensityData
    s['velocityd'] = arr_downstreamVelocityData
    s['fgmu'] = arr_upstreamFgmData
    s['densityu'] = arr_upstreamDensityData
    s['velocityu'] = arr_upstreamVelocityData
    s['ratio'] = arr_ratio
    s['G_u'] = arr_G_u
    s['G_d'] = arr_G_d

    s.close()



    
```
